```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline


x = np.array([0, 1, 2, 3])
y = np.exp(x)


f_prime_0 = 0
f_prime_n = 0


spline = CubicSpline(x, y, bc_type=((1, f_prime_0), (1, f_prime_n)))


x_vals = np.linspace(0, 3, 100)
y_vals = spline(x_vals)


plt.plot(x, y, 'o', label='Puntos de datos')
plt.plot(x_vals, y_vals, label='Spline c√∫bico interpolante')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.legend()
plt.show()
```
